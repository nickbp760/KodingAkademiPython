**DATA VISUALISATION**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
netflix = pd.read_csv('/content/netflix_titles-1.csv')

# Mengelompokkan data berdasarkan 'type' dan 'release_year'
grouped_netflix = netflix.groupby(['type', 'release_year']).size().reset_index(name='count')
print(grouped_netflix[['type', 'release_year', 'count']])

# Plot histogram for grouped data
plt.figure(figsize=(12, 8))

df = grouped_netflix[grouped_netflix['type'] == 'Movie']
plt.hist(df['release_year'], weights=df['count'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 1), label='Movie')

df = grouped_netflix[grouped_netflix['type'] == 'TV Show']
plt.hist(df['release_year'], weights=df['count'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 1), label='TV Show')

# # Create a histogram where the x-axis represents years and the bars are colored by 'type'
# for label, df in grouped_netflix.groupby('type'):
#     plt.hist(df['release_year'], weights=df['count'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 1), label=label)

plt.title('Distribusi Jumlah Entri Berdasarkan Tahun Rilis dan Type')
plt.xlabel('Tahun Rilis')
plt.ylabel('Jumlah Entri')
plt.legend(title='Type')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```
