# **DATA VISUALISATION**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
netflix = pd.read_csv('/content/netflix_titles-1.csv')

# Mengelompokkan data berdasarkan 'type' dan 'release_year'
grouped_netflix = netflix.groupby(['type', 'release_year']).size().reset_index(name='count')
print(grouped_netflix[['type', 'release_year', 'count']])

# Plot histogram for grouped data
plt.figure(figsize=(12, 8))

df = grouped_netflix[grouped_netflix['type'] == 'Movie']
plt.hist(df['release_year'], weights=df['count'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 1), label='Movie')

df = grouped_netflix[grouped_netflix['type'] == 'TV Show']
plt.hist(df['release_year'], weights=df['count'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 1), label='TV Show')

# # Create a histogram where the x-axis represents years and the bars are colored by 'type'
# for label, df in grouped_netflix.groupby('type'):
#     plt.hist(df['release_year'], weights=df['count'], bins=range(int(df['release_year'].min()), int(df['release_year'].max()) + 1), label=label)

plt.title('Distribusi Jumlah Entri Berdasarkan Tahun Rilis dan Type')
plt.xlabel('Tahun Rilis')
plt.ylabel('Jumlah Entri')
plt.legend(title='Type')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```

# **MINI PROJECT VISUALISATION**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
bali = pd.read_csv('/content/bali_tourist.csv')
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
bali_2019_1 = bali[bali['year'] == 2019][months].sum()
print(bali_2019_1)
plt.figure(figsize=(12, 8))
plt.hist(bali_2019_1.index, weights=bali_2019_1, bins=range(len(bali_2019_1) + 1), alpha=0.5)
plt.show()
```

```python
bali_2019_2 = bali[bali['year'] == 2019]
bali_2019_2 = bali_2019_2.groupby('area')[months].sum()
print(bali_2019_2)
plt.figure(figsize=(12, 8))
# Loop through each index (row) in the DataFrame
for idx in bali_2019_2.index:
    plt.hist(bali_2019_2.columns,
             weights=bali_2019_2.loc[idx],
             bins=range(len(bali_2019_2.columns) + 1),
             alpha=0.5, label=idx)
# Add labels and title
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.title('Monthly Distribution by Area')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
bali_filtered = bali[bali['year'].isin([2019, 2018])]
bali_filtered = bali_filtered.groupby('year')[months].sum()
print(bali_filtered)
plt.figure(figsize=(12, 8))
# Loop through each index (row) in the DataFrame for line plot
plt.plot(bali_filtered.columns, bali_filtered.loc[2018], marker='o', label='Year 2018', color='blue')

# Plot data for year 2019
plt.plot(bali_filtered.columns, bali_filtered.loc[2019], marker='o', label='Year 2019', color='red')
plt.show()
```
