# **DATA VISUALISATION**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
netflix = pd.read_csv('/content/netflix_titles-1.csv')

# Filter data for 'Movie' and 'TV Show'
movies = netflix[netflix['type'] == 'Movie']
print(movies)
tv_shows = netflix[netflix['type'] == 'TV Show']
print(tv_shows)

# # Plot histogram for 'Movie' and 'TV Show' release years
plt.figure(figsize=(12, 8))

# Plot the histogram for Movies
plt.hist(movies['release_year'], bins=range(int(movies['release_year'].min()), int(movies['release_year'].max()) + 10), alpha=0.6, label='Movie', color='blue', edgecolor='black')

# Plot the histogram for TV Shows
plt.hist(tv_shows['release_year'], bins=range(int(tv_shows['release_year'].min()), int(tv_shows['release_year'].max()) + 10), alpha=0.6, label='TV Show', color='orange', edgecolor='black')

# Add title and labels
plt.title('Distribusi Jumlah Entri Berdasarkan Tahun Rilis dan Type')
plt.xlabel('Tahun Rilis')
plt.ylabel('Jumlah Entri')

# Show legend
plt.legend(title='Type')

# Show grid and adjust x-axis ticks
plt.grid(axis='y')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

# **MINI PROJECT VISUALISATION**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
bali = pd.read_csv('/content/bali_tourist.csv')
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
bali_2019_1 = bali[bali['year'] == 2019][months].sum()
print(bali_2019_1)
plt.figure(figsize=(12, 8))
plt.hist(bali_2019_1.index, weights=bali_2019_1, bins=range(len(bali_2019_1) + 1), alpha=0.5)
plt.show()
```

```python
bali_2019_2 = bali[bali['year'] == 2019]
bali_2019_2 = bali_2019_2.groupby('area')[months].sum()
print(bali_2019_2)
plt.figure(figsize=(12, 8))
# Loop through each index (row) in the DataFrame
for idx in bali_2019_2.index:
    plt.hist(bali_2019_2.columns,
             weights=bali_2019_2.loc[idx],
             bins=range(len(bali_2019_2.columns) + 1),
             alpha=0.5, label=idx)
# Add labels and title
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.title('Monthly Distribution by Area')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
bali_filtered = bali[bali['year'].isin([2019, 2018])]
bali_filtered = bali_filtered.groupby('year')[months].sum()
print(bali_filtered)
plt.figure(figsize=(12, 8))
# Loop through each index (row) in the DataFrame for line plot
plt.plot(bali_filtered.columns, bali_filtered.loc[2018], marker='o', label='Year 2018', color='blue')

# Plot data for year 2019
plt.plot(bali_filtered.columns, bali_filtered.loc[2019], marker='o', label='Year 2019', color='red')
plt.show()
```
