```
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'id': [2019, 2019],
    'area': ['yay', 'bab'],
    'A': [10, 15],
    'B': [5, 10],
    'C': [3, 6]
}

# Create DataFrame
df = pd.DataFrame(data)
print(df)
yy_2019 = df.groupby('id')[['A', 'B', 'C']].sum()
print(yy_2019, type(yy_2019))
plt.hist('A', weights=yy_2019['A'])
plt.hist('B', weights=yy_2019['B'])
plt.hist('C', weights=yy_2019['C'])
plt.show()

# Show the plot
plt.hist(yy_2019.columns, weights=yy_2019.loc[2019])
plt.show()

# Filter for id 2019
df_2019 = df[df['id'] == 2019][['A', 'B', 'C']].sum()
print(df_2019, type(df_2019))
plt.hist(df_2019.index, weights=df_2019)
plt.show()

plt.hist(df_2019.index)
plt.show()
df_2019.hist()
plt.show()
```
