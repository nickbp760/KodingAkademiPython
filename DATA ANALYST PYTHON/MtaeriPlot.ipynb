```python
import pandas as pd
import matplotlib.pyplot as plt
# Membuat dataframe dengan kolom year, type, dan value
data = {
    "year": [2019, 2019, 2019, 2020, 2020, 2020],
    "type": ['A', 'B', 'C', 'A', 'B', 'C'],
    "value": [8, 13, 15, 20, 2, 6]
}

# pandas dataframe (baris dan kolom) -> pandas series (baris dan 1 kolom) -> numpy array/ndarray (baris horizontal tanpa kolom)
# tipe series yang dimasukan ke dalam index/weight plt.hist akan berubah auto ke ndarray(numpy array) dengan mengambil .values
# pada contoh berikut definisikan manual untuk tipe ndarray yang masuk ke plt hist(index=ndarray, weight=ndarray)
# pada plt hist, data dengan index yang sama maka akan auto ditambahkan

df = pd.DataFrame(data)
#bentuk data dataframe
print(df)
# bentuk data series
print(df['type'])
# bentuk data ndarray
print(type(df['type'].values), df['type'].values)
print(type(df['value'].values), df['value'].values)

# nilai dari x adalah ndarray, karena weight tidak define maka bernilai 1
# index akan digunakan sebagai nilai x
# weight akan digunakan sebagai nilai y
plt.hist(df['type'].values)
plt.show()

# nilai dari index adalah ndarray, karena weight terdefine maka bernilai sesuai data
# index akan digunakan sebagai nilai x
# weight akan digunakan sebagai nilai y
# INGAT JUMLAH DATA DARI X DAN Y HARUS SAMA
plt.hist(df['type'].values, weights=df['value'].values)
plt.show()

# plt hist yang berulang lalu plt show maka akan tampil pada canvas grafik yang sama
# misal disini kita pecah ke tahun yang berbeda
# biru = 2019, orange = 2020, coklat =2019
df_2019 = df[df['year'] == 2019]
df_2020 = df[df['year'] == 2020]
# Menggunakan bins yang lebih jelas agar tidak geser
plt.hist(df_2019['type'].values, weights=df_2019['value'].values, alpha=0.25, label='2019')
plt.hist(df_2020['type'].values, weights=df_2020['value'].values, alpha=0.25, label='2020')
plt.legend(['2019', '2020'])
plt.xlabel('Type')
plt.ylabel('Value')
plt.title('hist plot of Type and Value')
plt.show()
```
