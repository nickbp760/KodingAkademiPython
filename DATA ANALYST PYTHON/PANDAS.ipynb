**SERIES**

```python
import pandas as pd

# Membuat Series
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=index)

print("Series:")
print(series)
# Mengakses elemen dengan label 'c'
print("\nElemen dengan label 'c':")
print(series['c'])
# Mengakses elemen pada posisi ke-3 (indeks 2)
print("\nElemen pada posisi ke-3 (indeks 2):")
print(series[2])
# Memilih elemen dari 'b' hingga 'd' (inklusive)
print("\nSlicing dari 'b' hingga 'd':")
print(series['b':'d'])
# Memilih elemen dari posisi 1 hingga 3 (inklusif)
print("\nSlicing dari posisi 1 hingga 3:")
print(series[1:4])
# Memilih elemen yang lebih besar dari 30
print("\nElemen yang lebih besar dari 30:")
print(series[series > 30])
# Menggunakan .loc untuk memilih elemen dengan label 'a'
print("\nMenggunakan .loc untuk memilih elemen dengan label 'a':")
print(series.loc['a'])
# Menggunakan .iloc untuk memilih elemen pada posisi pertama
print("\nMenggunakan .iloc untuk memilih elemen pada posisi pertama:")
print(series.iloc[0])

```

```python
# Membuat Series dari dictionary
data_dict = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
series_from_dict = pd.Series(data_dict)

print("\nSeries dari dictionary:")
print(series_from_dict)

# Membuat Series dari skalar
scalar_value = 5
index = ['a', 'b', 'c', 'd']
series_from_scalar = pd.Series(scalar_value, index=index)

print("\nSeries dari skalar:")
print(series_from_scalar)

```

```python
import pandas as pd

# Membuat Series
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=index)

# Kondisi boolean dengan Series
bool_series = series > 20

# Menggunakan .loc dengan Series boolean
result_loc_series = series.loc[bool_series]

print("Hasil menggunakan .loc dengan Series boolean:")
print(result_loc_series)
# Kondisi boolean dengan list
bool_list = [False, False, True, True, False]

# Menggunakan .loc dengan list boolean
result_loc_list = series.loc[bool_list]

print("\nHasil menggunakan .loc dengan list boolean:")
print(result_loc_list)

# Kondisi boolean dengan list
bool_list = [True, False, True, False, True]

# Menggunakan .iloc dengan list boolean
result_iloc = series.iloc[bool_list]

print("\nHasil menggunakan .iloc dengan list boolean:")
print(result_iloc)
```

```python
import pandas as pd

# Membuat Series
series1 = pd.Series([10, 20, 30, 40])
series2 = pd.Series([1, 2, 3, 4])

# Penjumlahan
print("Penjumlahan:")
print(series1 + series2)

# Pengurangan
print("\nPengurangan:")
print(series1 - series2)

# Perkalian
print("\nPerkalian:")
print(series1 * series2)

# Pembagian
print("\nPembagian:")
print(series1 / series2)

# Statistik Deskriptif
print("Rata-rata:")
print(series1.mean())

print("\nMedian:")
print(series1.median())

print("\nStandar Deviasi:")
print(series1.std())

print("\nDeskripsi:")
print(series1.describe())
```

**DATA FRAME**

```python
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}

# Create DataFrame
df = pd.DataFrame(data)
print(df)
# Set 'Name' as the index
df.set_index('Name', inplace=True)
print("columns")
print(df.columns)
print("indexs")
print(df.index)
# Mencari indeks dari label 'Charlie'
index_position = df.index.get_loc('Charlie')
print("\nPosisi indeks untuk label 'Charlie' adalah:", index_position)
print("\nset Name as an Index")
print(df)
print("\nselect Single Column")
print(df['Age'])
print("\nselect Mutli Column")
print(df[['Age', 'City']])
print("\nselect Single Row")
print(df.loc['Alice'])
print("\nselect Multi Row")
print(df.loc[['Alice', 'Eva']])
print("\nselect Index Range Row")
print(df.loc['Alice':'Charlie'])
print("\nselect By Conditions")
print(df.loc[df['Age'] > 25, 'Age'])
df.reset_index(inplace=True)
# Pilih baris kedua dan ketiga (indeks 1 dan 2) dan kolom pertama dan kedua (indeks 0 dan 1)
selection = df.iloc[1:3, 0:2]
print("\ncontoh pakai iloc RANGE")
print(selection)
# Pilih baris pertama dan ketiga (indeks 0 dan 2) dan kolom indeks ke 1
selected_rows = df.iloc[[0, 2], [1]]
print("\ncontoh pakai iloc milih kolom baris tertentu")
print(selected_rows)
```

```python
import pandas as pd

# Data contoh
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Occupation': ['Engineer', 'Doctor', 'Artist', 'Lawyer', 'Scientist'],
    'Salary': [70000, 80000, 50000, 120000, 95000]
}

# Membuat DataFrame
df = pd.DataFrame(data)

df = pd.DataFrame(data)
df.set_index('Name', inplace=True)
print("df lengkap")
print(df)
a = df.drop("Alice")
print("\ndf drop Alice")
print(a)
b = df.drop(["Bob", "Charlie"])
print("\ndf drop Bob Charlie")
print(b)
c = df.drop(columns=["City", "Occupation"])
print("\ndf drop column City Occupation")
print(c)
d = df[['Salary', 'Age']] - 1
print("\nSalary Age -1")
print(d)
hh = pd.Series([1000,3], index=['Salary', 'Age'])
print("\ndf Salary Age dari Series")
print(hh)
print(df[['Salary', 'Age']]-hh)
langs = pd.Series(["nn", "bb", "cc"], index=["Alice", "Bob", "David"], name = "random")
df["random"] = langs
print("\ndf Series Random")
print(df)
df["random"] = "ok"
print("\ndf Series No Random")
print(df)
df["capacity"] = df["Salary"] / df["Age"]
print("\ndf capacity")
print(df)
rename = df.rename(
    columns={
        'Salary' : "gaji",
        'Age' : "umur"
    },
    index={
        'Bobb' : "Bobby",
        'Alice' : "Alice"
    }
)
print("\ndf rename")
print(rename)
```
